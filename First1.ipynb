Q.1. What are sequence data types in Python?
Name them and explain how indexing and slicing work across different sequence types?

Answer:

Sequence Data Types

Sequence data types are those that store ordered collections of elements.

Main sequence types:
List,
Tuple,
String,
Range

Indexing

Indexing means accessing elements using their position.

Index starts from 0.

Negative indexing is also allowed (from end).

e.g.
data = [10, 20, 30]

print(data[0])   # 10

print(data[-1])  # 30

Slicing

Slicing is used to extract a part of sequence.

Syntax:

sequence[start:end:step]

e.g.

print(data[0:2])  # [10, 20]

Q.2. What is the difference between a list and a tuple in Python?
Why would you prefer a tuple over a list in real-world applications?

Answer:

List

Mutable (can be changed)

Defined using []

Slower compared to tuple

Used for dynamic data

Tuple

Immutable (cannot be changed)

Defined using ()

Faster and memory efficient

Used for fixed data

Why prefer Tuple in real-world applications?

Data security (cannot be modified)

Faster performance

Used for fixed values like coordinates, database records

Q.4. Explain how dictionaries work internally in Python.
Why are dictionary lookups faster compared to lists or tuples?

Answer:

Dictionary

A dictionary stores data in key-value pairs.

Internal Working

Uses a hash table internally

Each key is converted into a hash value

This hash points directly to the value location

Why lookups are fast?

Direct access using hash function

No need to search entire structure

Average time complexity is O(1)

Compared to Lists/Tuples

Lists/Tuples require linear search (O(n))

Dictionaries provide direct access

Q.5. What is the difference between sets and lists in Python?
How do sets handle duplicate values, and where are sets useful in practical scenarios?

Answer:

List
Ordered collection

Allows duplicates

Indexed

Set

Unordered collection

Does NOT allow duplicates

No indexing

Automatically removes duplicate values
e.g.

s = {1, 2, 2, 3}

print(s)   # {1, 2, 3}

Practical Uses of Sets

Removing duplicate data

Finding unique elements

Mathematical operations (union, intersection)

Fast membership testing

```python
import array as arr
array=arr.array('i',[10,20,30])
print(array)
array.insert(1,15)
print(array)
array.remove(20)
print(array)
print("Traversing Elements:")
for i in array:
  print(i)
```

```python
print("\n TUPLE")
t=(100,200,300)
print(t[0])
print(t[0:2])
print(len(t))
print(t.count(200))
```

```python
print("\n DICTIONARY")
d={"name":"GOPAL","age":21}
print(d["name"])
d["city"]="Mumbai"
print(d)
d["age"]=22
print(d)
del d["city"]
print(d)
```

```python
print("\n SET")
set1={1,2,3}
set2={4,5,6}
set1.add(6)
print(set1)
set1.remove(2)
print(set1)
print(set1.union(set2))
print(set1.intersection(set2))
```
