Q.1What is the difference between Binomial, Poisson, and Normal distributions?

Answer:

Binomial Distribution:

The Binomial distribution is a discrete probability distribution used when there is a fixed number of trials.

In this distribution, each trial has only two possible outcomes: success or failure.

It is defined by two parameters: n (number of trials) and p (probability of success).

e.g.the Binomial distribution is tossing a coin multiple times and counting the number of heads.


Poisson distribution:

The Poisson distribution is also a discrete probability distribution used to count the number of events occurring in a fixed interval of time or space.

It is suitable for modeling rare and independent events.

It is defined by a single parameter λ (lambda), which represents the average rate of occurrence.

e.g. the number of calls received per hour.

Normal distribution:
The Normal distribution is a continuous probability distribution that has a bell-shaped curve.

It is defined by two parameters: μ (mean) and σ (standard deviation).

It is commonly used to represent natural and continuous data such as height and weight.

Q.3.What is the Central Limit Theorem and how does it relate to Normal distribution?

Answer:

The Central Limit Theorem (CLT) is an important concept in statistics that explains how sample means behave.

It states that when we take a large number of samples from any population and calculate their means, the distribution of those sample means becomes approximately Normal.

This is true even if the original data is not normally distributed and may be skewed or irregular.

$\bar{X}$∼N( μ ,​σ/$\sqrt{n}$

)

According to CLT, the mean of the sample means is equal to the population mean (μ).

The standard deviation of the sample means is σ/√n, which is called the standard error.

As the sample size increases, the distribution becomes more smooth, symmetric, and closer to a bell-shaped Normal curve.

This explains why the Normal distribution appears frequently in real-world data and statistical analysis.

The CLT allows statisticians to use Normal distribution methods for hypothesis testing, confidence intervals, and predictions, even when the original data is not Normal.

Overall, the Central Limit Theorem provides the foundation for applying the Normal distribution in statistics and machine learning.

Q.5.What is skewness in distributions and which distributions exhibit it?

Answer:

Skewness is a measure of the asymmetry of a probability distribution.

It shows how much the data is tilted to one side compared to a symmetric distribution.

There are three types of skewness:
Positive skewness (right-skewed): The tail of the distribution is longer on the right side.

Negative skewness (left-skewed): The tail is longer on the left side.

Zero skewness: The distribution is perfectly symmetric.

The Normal distribution has zero skewness, as it is perfectly symmetric.

The Poisson distribution is generally positively skewed, especially when the value of λ is small.

The Binomial distribution can be symmetric or skewed depending on the probability value (p).

It is symmetric when p = 0.5
It becomes skewed when p ≠ 0.5

Skewness is important because it helps in understanding the shape of the data and selecting appropriate statistical methods.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Create figure and subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("Analysis of Binomial, Poisson, and Normal Distributions", fontsize=16)

# ------------------- BINOMIAL DISTRIBUTION -------------------

n, p = 10, 0.5

x_binom = np.arange(0, n + 1)

pmf_binom = binom.pmf(x_binom, n, p)
cdf_binom = binom.cdf(x_binom, n, p)

samples_binom = binom.rvs(n, p, size=1000)

# PMF Graph
axs[0, 0].bar(x_binom, pmf_binom, color='skyblue')
axs[0, 0].set_title("Binomial PMF (n=10, p=0.5)")

# CDF Graph
axs[0, 1].step(x_binom, cdf_binom, where='mid', color='blue')
axs[0, 1].set_title("Binomial CDF")

# Histogram
axs[0, 2].hist(samples_binom, bins=np.arange(n + 2) - 0.5, density=True,color='navy', alpha=0.7)

axs[0, 2].set_title("Binomial Histogram")

# ------------------- POISSON DISTRIBUTION -------------------

mu = 4

x_poisson = np.arange(0, 15)

pmf_poisson = poisson.pmf(x_poisson, mu)
cdf_poisson = poisson.cdf(x_poisson, mu)

samples_poisson = poisson.rvs(mu, size=1000)

# PMF Graph
axs[1, 0].bar(x_poisson, pmf_poisson, color='lightgreen')
axs[1, 0].set_title("Poisson PMF (μ=4)")

# CDF Graph
axs[1, 1].step(x_poisson, cdf_poisson, where='mid', color='green')
axs[1, 1].set_title("Poisson CDF")

# Histogram
axs[1, 2].hist(samples_poisson, bins=np.arange(16) - 0.5,  density=True, color='darkgreen', alpha=0.7)

axs[1, 2].set_title("Poisson Histogram")

# ------------------- NORMAL DISTRIBUTION -------------------

mu_norm = 0
sigma = 1

x_norm = np.linspace(-4, 4, 1000)

pdf_norm = norm.pdf(x_norm, mu_norm, sigma)
cdf_norm = norm.cdf(x_norm, mu_norm, sigma)

samples_norm = norm.rvs(mu_norm, sigma, size=1000)

# PDF Graph
axs[2, 0].plot(x_norm, pdf_norm, color='salmon')
axs[2, 0].set_title("Normal PDF (μ=0, σ=1)")

# CDF Graph
axs[2, 1].plot(x_norm, cdf_norm, color='red')
axs[2, 1].set_title("Normal CDF")

# Histogram
axs[2, 2].hist(samples_norm, bins=30, density=True, color='darkred', alpha=0.7)

axs[2, 2].set_title("Normal Histogram")

# Add grid to all graphs
for row in axs:
    for ax in row:
        ax.grid(True)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Show graphs
plt.show()

# ------------------- STATISTICAL ANALYSIS -------------------

print("Binomial Distribution:")
print(f"Mean = {np.mean(samples_binom):.2f}")
print(f"Variance = {np.var(samples_binom):.2f}")

print("\nPoisson Distribution:")
print(f"Mean = {np.mean(samples_poisson):.2f}")
print(f"Variance = {np.var(samples_poisson):.2f}")

print("\nNormal Distribution:")
print(f"Mean = {np.mean(samples_norm):.2f}")
print(f"Standard Deviation = {np.std(samples_norm):.2f}")
```
