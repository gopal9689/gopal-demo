Q.1.	What is data wrangling? Explain the difference between data wrangling and data preprocessing.

Answer:

Data Wrangling

Data wrangling is the process of cleaning, transforming, and organizing raw data into a structured and usable format for analysis.

Difference:

Data Wrangling

It is used to clean and organize raw data

It focuses on handling missing values, removing errors, and structuring data

It is performed in the initial stage of data analysis

Common tasks include removing duplicates, fixing formats, filtering data

The main goal is to make data clean and usable

Data Preprocessing

It is used to prepare data for machine learning models

It focuses on scaling, encoding, and transforming features

It is performed after data wrangling and before model building

Common tasks include normalization, standardization, encoding categorical values

The main goal is to make data suitable for algorithms

Q.3.	What are the main differences between Series, DataFrame, and Panel?

Answer:

Series

It is a one-dimensional (1D) array

Stores single column of data

Contains values with index labels

Can store different data types

Used for handling a single variable


DataFrame

It is a two-dimensional (2D) data structure

Stores data in rows and columns (table format)

Each column can have different data types

Most commonly used structure in Pandas

Used for tabular data analysis

Panel

It is a three-dimensional (3D) data structure

Stores data as a collection of DataFrames

Used for multi-dimensional data

Deprecated in modern Pandas (not used now)

```python
import pandas as pd
# Step 1: Create dataset (no need to upload file)
data = {
    'Name': ['Amit','Neha','Ravi','Priya','Amit','Neha','Ravi','Priya'],
    'Subject': ['Math','Math','Science','Science','Math','Math','Science','Science'],
    'Semester': [1,1,1,1,2,2,2,2],
    'Marks': [78,85,82,88,81,79,85,90]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
#  BASIC OPERATIONS
# 1. Select a single column
print("\n1. Selecting 'Marks' column:")
print(df['Marks'])
# 2. Add a new column with constant values
df['College'] = "ABC University"
print("\n2. After adding new column 'College':")
print(df)
# 3. Sort DataFrame by one column
sorted_df = df.sort_values(by='Marks', ascending=False)
print("\n3. Sorted DataFrame by Marks:")
print(sorted_df)
# 4. Rename DataFrame columns
df = df.rename(columns={'Marks': 'Score'})
print("\n4. After renaming 'Marks' to 'Score':")
print(df)
# 5. Display summary statistics
print("\n5. Summary Statistics:")
print(df.describe())
#  DATA WRANGLING (MAIN TASK)
# Group by Subject and Semester and calculate average
grouped = df.groupby(['Subject','Semester'])['Score'].mean()
print("\nGrouped Average (Subject & Semester):")
print(grouped)
# Convert to DataFrame
summary = grouped.reset_index()
print("\nFinal Summary DataFrame:")
print(summary)
# Pivot Table
pivot = df.pivot_table(values='Score', index='Subject', columns='Semester', aggfunc='mean')
print("\nPivot Table:")
print(pivot)
# Save result
summary.to_csv('final_summary.csv', index=False)
```
