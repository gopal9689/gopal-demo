Q.3. Difference between Web Scraping and Web Crawling .
Web Scraping

Answer:

Web scraping is the process of extracting specific data from websites. It focuses on collecting useful information such as text, images, prices, tables, or links from a webpage.

It targets particular data elements
Uses tools like Python libraries (Requests, BeautifulSoup)
Works on already known URLs
Example: Extracting product prices from an e-commerce website
Web Crawling

Web crawling is the process of automatically browsing and discovering web pages by following links from one page to another.

It focuses on finding and indexing web pages
Used by search engines
Starts from a seed URL and follows hyperlinks
Example: Search engines scanning websites to index pages

Q.4. Commonly Used Methods in BeautifulSoup (Detailed)

Answer:

BeautifulSoup provides several methods to parse and extract data from HTML documents easily. These methods help in navigating, searching, and modifying the parse tree.

1. find()

Returns the first matching HTML element

Useful when only one element is needed

Example:

soup.find("h1")

2. find_all()

Returns all matching elements as a list

Most commonly used method

Example:

soup.find_all("p")

3. get_text()

Extracts text content from a tag

Removes HTML tags

Example:

soup.find("p").get_text()

4. get()

Used to get attribute values of a tag (like href, src)

Example:

link.get("href")

5. select()

Uses CSS selectors to find elements

More flexible and powerful

Example:

soup.select("div.classname")

6. select_one()

Returns the first element matching a CSS selector

Example:

soup.select_one("#idname")

7. prettify()

Formats the HTML document in a readable structure

Example:

print(soup.prettify())

8. find_parent() / find_parents()

Finds the parent element(s) of a tag

Example:

tag.find_parent("div")

9. find_next() / find_next_sibling()

Finds the next element or sibling in the HTML structure

Example:

tag.find_next("p")

10. attrs

Accesses all attributes of a tag as a dictionary

Example:

tag.attrs

Q.5. Ethical and Legal Considerations in Web Scraping.

Answer:Ethical Considerations

1. Respect Website Terms and Conditions

Every website has its own Terms of Service (ToS)

Always read and follow them before scraping

Some websites strictly prohibit scraping

2. Follow robots.txt Rules

Websites provide a robots.txt file that specifies which parts can be accessed by bots
Scrapers should respect these rules

3. Avoid Overloading Servers

Sending too many requests in a short time can crash or slow down a website
Use delays (time.sleep()) between requests

4. Do Not Extract Sensitive Data

Avoid scraping personal, private, or confidential information
Respect user privacy
5. Give Credit When Required

If using scraped data publicly, mention the source
Avoid plagiarism

 Legal Considerations

1. Copyright Laws

Website content (text, images, videos) may be copyrighted

Do not reuse or distribute without permission

2. Data Protection Laws

Laws like GDPR (Europe) protect user data

Avoid collecting personal information without consent

3. Website Restrictions

Some websites legally restrict automated access

Violating these rules can lead to penalties or bans

4. Intellectual Property Rights

Respect ownership of data and content

Do not misuse proprietary data

5. Risk of Legal Action

Unauthorized scraping may lead to:

Account blocking

IP banning

Legal notices or lawsuits

```python
import requests
from bs4 import BeautifulSoup
import csv
# Step 1: Define URL
url = "https://books.toscrape.com/"
# Step 2: Send HTTP GET request
response = requests.get(url)
# Step 3: Check status
if response.status_code == 200:
    print("Successfully fetched the webpage")
else:
    print(f"Failed. Status code: {response.status_code}")
    exit()
# Step 4: Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')
# Step 5: Find 5 book containers
# Each book is inside <article class="product_pod">
books = soup.find_all('article', class_='product_pod')[:5]
print(f"\nTotal books found: {len(books)}")
print("=" * 55)
# Step 6: Extract data from each book
data = []
for index, book in enumerate(books, start=1):
    # Extract BOOK TITLE
    title = book.find('h3').find('a')['title']
    # Extract PRICE
    price = book.find('p', class_='price_color').get_text()
    # Extract RATING
    rating_tag = book.find('p', class_='star-rating')
    rating = rating_tag['class'][1] # Gets: One, Two, Three, Four, Five
    # Extract AVAILABILITY
    availability = book.find('p', class_='instock availability').get_text(strip=True)
    print(f"\nBook {index}:")
    print(f" Title : {title}")
    print(f" Price : {price}")
    print(f" Rating : {rating} Star")
    print(f" Availability : {availability}")
    data.append({
        'Book No' : index,
        'Title' : title,
        'Price' : price,
        'Rating' : rating + " Star",
        'Availability' : availability
    })
# Step 7: Save to CSV
filename = "books.csv"
with open(filename, mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ['Book No', 'Title', 'Price', 'Rating', 'Availability']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for row in data:
        writer.writerow(row)
    print(f"\n{'=' * 55}")
    print(f"Data saved to '{filename}' successfully!")
    print(f"Total books saved: {len(data)}")
# Step 8: Read back CSV to verify
print("\n── Verifying CSV Contents ──")
print("=" * 55)
with open(filename, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(f"\nBook {row['Book No']}:")
        print(f" Title : {row['Title']}")
        print(f" Price : {row['Price']}")
        print(f" Rating : {row['Rating']}")
        print(f" Availability : {row['Availability']}")
```

```python
# ── PART 2: Generic Scraper (User Input) + Save to CSV ──────
import requests
from bs4 import BeautifulSoup
import csv

# Step 1: Take inputs from user
url = input("Enter the URL to scrape: ")
tag = input("Enter HTML tag (e.g., p, h1, div): ")
class_name = input("Enter class name (blank if none): ")

# Define a User-Agent header to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Step 2: Fetch the webpage with error handling
try:
    response = requests.get(url, headers=headers, timeout=10)
    if response.status_code != 200:
        print("Failed. Status code:", response.status_code)
        exit()
except Exception as e:
    print("Error:", e)
    exit()

# Step 3: Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Step 4: Find elements — with class or without
if class_name.strip():
    elements = soup.find_all(tag, class_=class_name.strip())
else:
    elements = soup.find_all(tag)

# Step 5: Display results + store in list
data = [] # Initialize data list here to ensure it's empty for each run
if not elements:
    print("No elements found.")
else:
    print(f"\nFound {len(elements)} element(s):\n")
    for i, el in enumerate(elements, 1):
        text = el.get_text(strip=True)
        attrs = str(el.attrs)
        print(f"Element {i}:")
        print(" Text :", text)
        print(" Attributes:", attrs)
        print("-" * 50)
        data.append({
            'Element No' : i,
            'Tag' : tag,
            'Text' : text,
            'Attributes' : attrs
        })

# Step 6: Print page title
title = soup.title.string if soup.title else "N/A"
print("\nPage Title:", title)

# Step 7: Save to CSV file
if data:
    filename = "scraped_data.csv"
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        fieldnames = ['Element No', 'Tag', 'Text', 'Attributes']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for row in data:
            writer.writerow(row)
    print(f"\nData saved to '{filename}' successfully!")
    print(f"Total rows written: {len(data)}")

    # Step 8: Read back CSV to verify - only if data was actually written
    print("\n── Verifying CSV Contents ──")
    print("=" * 50)
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            print(f"\nElement {row['Element No']}:")
            print(f" Tag : {row['Tag']}")
            print(f" Text : {row['Text']}")
            print(f" Attributes : {row['Attributes']}")
else:
    print("\nNo data was scraped, so no CSV file was created or verified.")
```
