Q.1. What is Data Ingestion and Why is it Important in Data Science? Which Python library is commonly used?
Data Ingestion

Answer:

Data ingestion is the process of collecting and importing data from different sources into a system where it can be stored, processed, and analyzed.

Importance

Brings data from multiple sources (CSV, databases, APIs, Excel, etc.)

Makes data available for analysis and machine learning

Ensures data is organized in a usable format

Acts as the first step in the data pipeline

Common Python Library

The most commonly used library for data ingestion and cleaning is Pandas

Q.2.Why is Data Cleaning Necessary Before Analysis or Machine Learning?

Answer:

Data cleaning is important because raw data is often incomplete, inconsistent, or incorrect.

Reasons:

Fixes errors and inconsistencies

Improves data quality

Increases accuracy of analysis and models

Prevents misleading results

Q.4.What is the difference between head() and tail() functions in Pandas?

Answer:

head()

Displays the first n rows of the dataset by default

Used to quickly check initial data

e.g.df.head()

tail()

Displays the last n rows of the dataset by default

Used to check ending data

e.g.df.tail()

Q.5. How can you Identify Missing Values in a Dataset?

Answer:

Missing values can be identified using Pandas functions:

1. isnull()

Shows missing values as True

e.g.df.isnull()

2. isnull().sum()

Shows total missing values in each column

e.g.df.isnull().sum()

3. info()

Gives summary including non-null counts

e.g.df.info()

```python
import pandas as pd
# 1. Load CSV file
df = pd.read_csv("employee_performance_data.csv")
# Display first five rows
print("First 5 rows:")
print(df.head())
# 2. Identify missing values
print("\nMissing values:")
print(df.isnull())

# Count missing values
print("\nMissing values count:")
print(df.isnull().sum())
# 3. Replace missing values with 0
df_filled = df.fillna(0)
print("\nAfter replacing missing values with 0:")
print(df_filled)
# 4. Rename column names to lowercase
df_filled.columns = df_filled.columns.str.lower()
print("\nColumn names in lowercase:")
print(df_filled.columns)
# 5. Display dataset shape
print("\nDataset shape (rows, columns):")
print(df_filled.shape)
```
